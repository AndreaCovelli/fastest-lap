```python
# Load vehicle
vehicle=fastest_lap.load_vehicle("car","limebeer-2014-f1","../../../../database/limebeer-2014-f1.xml");
```

```python
# Compute and plot gg_diagram
[ay,ay_minus,ax_max,ax_min] = fastest_lap.gg_diagram(vehicle,250.0*KMH,100);
```
